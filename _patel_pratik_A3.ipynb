```python
import pandas as pd

uncleaned_df = pd.read_csv("UnCleaned_Conestoga_Dataset.csv")

# Display the first few rows of each dataset
print(uncleaned_df.head())

#INFO
print(uncleaned_df.info())
```

# data cleaning

```python
# Extract feet and inches from the 'Height' column using str.extract function
uncleaned_df['FeetTemp'] = uncleaned_df['Height'].str.extract(r"(\d)'").astype(float)
uncleaned_df['InchTemp'] = uncleaned_df['Height'].str.extract(r"'(\d+)").astype(float)

# Convert to centimeters
uncleaned_df['Height(cm)'] = (uncleaned_df['FeetTemp'] * 30.48 + uncleaned_df['InchTemp'] * 2.54).round(2)

# Droping the temporary columns because we already converted it and using inplace to assign the value into variable
uncleaned_df.drop(columns=['FeetTemp', 'InchTemp'], inplace=True)
uncleaned_df
```

```python
# Remove 'lbs' and convert to numeric
uncleaned_df['Weight(lbs)'] = uncleaned_df['Weight'].str.replace('lbs', '').astype(float)

# Verifying the the changes by using two columns and head function
print(uncleaned_df[['Weight', 'Weight(lbs)']].head())
```

```python
# Split the 'JoinedCollege' column into day, month, and year using the same split function
uncleaned_df[['Day Joined', 'Month Joined', 'Year Joined']] = uncleaned_df['JoinedCollege'].str.split('-', expand=True)

# Verify the split
print(uncleaned_df[['JoinedCollege', 'Day Joined', 'Month Joined', 'Year Joined']].head())

uncleaned_df
```

```python
# Split the 'car they own' column into carmake, and car model using the split function
uncleaned_df['CarTheyOwn'] = uncleaned_df['CarTheyOwn'].astype(str).str.replace('/', ' ')
uncleaned_df[['Car Make', 'Car Model']] = uncleaned_df['CarTheyOwn'].str.split(' ',n=1, expand=True)


uncleaned_df



```

```python
# Split the 'personal status' column into gender, and status using the split function
uncleaned_df['personal_status'] = uncleaned_df['personal_status'].str.replace('/', ' ')

# Split 'personal_status' into two columns 'gender' and 'status' with correct syntax
uncleaned_df[['Gender', 'Status']] = uncleaned_df['personal_status'].str.split(' ', n=1, expand=True)

# Verify the changes
uncleaned_df
```

```python
#Filling missing values

# Fill 'Born' with the median
uncleaned_df['Born'] = uncleaned_df['Born'].fillna(uncleaned_df['Born'].median())

# Fill 'HomeTown' with 'Unknown'
uncleaned_df['HomeTown'] = uncleaned_df['HomeTown'].fillna('Unknown')

# Fill 'personal_status' and 'job' with their respective modes
uncleaned_df['personal_status'] = uncleaned_df['personal_status'].fillna(uncleaned_df['personal_status'].mode()[0])
uncleaned_df['job'] = uncleaned_df['job'].fillna(uncleaned_df['job'].mode()[0])

# Fill 'Favorite App' with 'Unknown'
uncleaned_df['Favorite App'] = uncleaned_df['Favorite App'].fillna('Unknown')

# Verify the changes
uncleaned_df.info()
```

```python
# Drop the 'HomeTown' and 'Favorite App' columns from the DataFrame
uncleaned_df = uncleaned_df.drop(['HomeTown', 'Favorite App','job'], axis=1)

# Verify the result by checking the columns
print(uncleaned_df.columns)


```

# data visualisation


```python
import seaborn as sns
import matplotlib.pyplot as plt
###Does age correlate with credit amount, and if so, how?

# Scatter plot for Age vs. Credit_Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x=uncleaned_df['Age'], y=uncleaned_df['credit_amount'], color='blue', alpha=0.7)

# Adding the titles and labels
plt.title("Scatter Plot: Age vs. Credit Amount", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("credit_amount ($)", fontsize=12)

# Displaying thr  correlation coefficient between age and credit amount
correlation = uncleaned_df[['Age', 'credit_amount']].corr().iloc[0, 1]
print(f"Correlation Coefficient (Age vs. Credit_Amount): {correlation:.2f}")

plt.show()




```

```python
import matplotlib.pyplot as plt
import seaborn as sns
###Identify the top 10 female students with the highest weights(lbs)
# Filtering the dataset for female students
female_students = uncleaned_df[uncleaned_df['Gender'] == 'female']

# Sort by weight in descending order to put higher value first on top
female_students_sorted = female_students.sort_values(by='Weight', ascending=False)

# Select the top 10 female students with highest weights by using head function
top_10_female_students = female_students_sorted.head(10)

# Create a bar plot to visualize the top 10 female students by weight
plt.figure(figsize=(10, 6))
sns.barplot(x='Weight', y='Name', data=top_10_female_students, palette='Blues_d')

# Adding labels and titles
plt.title('Top 10 Female Students with the Highest Weights')
plt.xlabel('Weight (lbs)')
plt.ylabel('Student Name')

# Show the plot
plt.show()

```
